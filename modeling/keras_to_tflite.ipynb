```python
import tensorflow as tf
from keras.models import load_model

# 1. Load model Keras yang sudah dilatih
model_path = 'ResNet50V2.keras'
model = load_model(model_path)

# 2. Konversi model Keras ke format TensorFlow Lite
converter = tf.lite.TFLiteConverter.from_keras_model(model)

# 3. (Opsional) Terapkan optimasi untuk ukuran model yang lebih kecil
# Misalnya, menggunakan optimasi ukuran (quantization)
converter.optimizations = [tf.lite.Optimize.DEFAULT]

# 4. Konversi model ke format .tflite
tflite_model = converter.convert()

# 5. Simpan model TFLite ke file
tflite_model_path = 'model.tflite'
with open(tflite_model_path, 'wb') as f:
    f.write(tflite_model)

print(f'Model TFLite telah disimpan di {tflite_model_path}')
```
